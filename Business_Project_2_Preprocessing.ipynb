```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras 
from sklearn.preprocessing import MinMaxScaler
sns.set()
```

```python
raw_data=pd.read_csv('dd.csv')
df=raw_data.copy()
```

```python
df.head(3)
```

```python
#droping the ID 
df=df.drop(columns=['ID'])
```

```python
reasons=pd.get_dummies(df['Reason for Absence'])
reasons
```

```python
reasons['SUM']=reasons.sum(axis=1)
```

```python
reasons=reasons.drop(columns=[0,'SUM'])
```

```python
# Grouping the reason to decreas the features
group_1=reasons.iloc[:, :15].max(axis=1)
group_2=reasons.iloc[:,15:18].max(axis=1)
group_3=reasons.iloc[:,18:22].max(axis=1)
group_4=reasons.iloc[:,22:].max(axis=1)
```

```python
df=pd.concat([df,group_4,group_3,group_2,group_1], axis=1 )
```

```python
df.columns.values
```

```python
# Renaming the last 3 columns
new_columns=['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Group_4', 'Group_3', 'Group_2', 'Group_1']
df.columns=new_columns
df.head(3)
```

```python
reorganized_columns=['Group_4', 'Group_3', 'Group_2', 'Group_1','Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
df=df[reorganized_columns]
```

```python
df.head(3)
```

```python
# Check Point (saving as CSV)
df_ch1=df.copy()
df.to_csv('Check_Point_1.csv', index=False)
```

```python
df2=df
```

```python
df2['Date']=pd.to_datetime(df2['Date'], format='%d/%m/%Y')
```

```python
df2['Date'].shape[0]
```

```python
month=[]
for i in range(df2['Date'].shape[0]):
    month.append(df2['Date'][i].month)
```

```python
df2['Month_']=month
```

```python
def date_to_weekday(date):
    return date.weekday()
```

```python
df2['WeekDay']=df2['Date'].apply(date_to_weekday)
```

```python
df2['Education']=df2['Education'].map({1:0,2:1,3:2,4:3})
```

```python
df2.to_csv('Preprocessed.csv', index=False)
```
